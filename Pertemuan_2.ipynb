```python
t=np.arange(0.0,5.0,0.2)#daftar nilai x= 0 sampai 5
y1=t
y2=t**2
y3=t**3

plt.plot(t,y1,'r--',t,y2,'bs',t,y3,'g^')
#r-- red strip 

plt.show()
```

```python
def f(t):
    return np.exp(-t) * np.cos(2 * np.pi * t)

t1=np.arange(0.0,5.0,0.1)
t2=np.arange(0.0,5.0,0.02)

y1=f(t1)
y2=np.cos(2 * np.pi *t2)
```

```python
plt.figure(dpi=300)

plt.subplot(211)
plt.plot(t1,y1,'bo',t2,f(t2),'k')
#k black

plt.subplot(212)
plt.plot(t2,y2,'r--')

plt.savefig('example.png')
plt.show()
```
