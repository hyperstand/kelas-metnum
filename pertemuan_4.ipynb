```python
from __future__ import division
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
f = lambda x : x * np.exp(-x / 2) - 0.5
```

```python
def hitung_error(nilai_sekarang,nilai_sebelumnya):
    return abs(nilai_sekarang - nilai_sebelumnya)/nilai_sekarang * 100
```

```python
def biseksi(f,xi,xu,es):
    aHasil=[]
    aError=[]
    error=100
    i=0
    
    while (error > es):
        xr = (xi+xu) / 2
        aHasil.append(xr)
        
        if(i!=0):
            error=hitung_error(aHasil[i],aHasil[i-1])
            aError.append(error)
        if(f(xi) * f(xr) < 0):
            xu=xr
        elif( f(xi) * f(xr) > 0):
            xi=xr
        else:
            break
        
        i += 1
    
    return (aHasil,aError,i)
```

```python
x=np.arange(0.0,1.0,0.01)
y=f(x)
plt.plot(x,y)
plt.plot(x,[0]*len(x),'r--')
plt.show()
```

```python
(hasil,error,i)=biseksi(f,0,1,5)
print (hasil)
print (error)

plt.subplot(211)
# plt.plot(range(1,i+1),hasil,'bo-')
plt.plot(range(1,len(hasil)+1),hasil,'bo-')
plt.ylabel("Akar xr")

plt.subplot(212)
# plt.plot(range(2,i+1),error,'rs-')
plt.plot(range(2,len(hasil)+1),error,'rs-')
plt.xlabel("Iterasi")
plt.ylabel("Error (%)")


```

```python
def Regulafalsi(f,xi,xu,es):
    aHasil=[]
    aError=[]
    error=100
    i=0
    
    while (error > es):
 
        xr=xu-(f(xu)*(xu-xi))/(f(xu)-f(xi));
        aHasil.append(xr)
        
        if(i!=0):
            error=hitung_error(aHasil[i],aHasil[i-1])
            aError.append(error)
        if(f(xi) * f(xr) < 0):
            xu=xr
        elif( f(xi) * f(xr) > 0):
            xi=xr
        else:
            break
        
        i += 1
    
    return (aHasil,aError,i)
```

```python
(hasil,error,i)=Regulafalsi(f,0,1,0.01)
print (hasil)
print (error)

plt.subplot(211)
plt.plot(range(1,len(hasil)+1),hasil,'bo-')
plt.ylabel("Akar xr")

plt.subplot(212)
# plt.plot(range(2,i+1),error,'rs-')
plt.plot(range(2,len(hasil)+1),error,'rs-')
plt.xlabel("Iterasi")
plt.ylabel("Error (%)")
```
