## Menghitung deret Mac Laurin untuk ${e^x}$


$$
e^x= \sum_{n=0}^{\tilde} \frac{x^n}{n!}
$$

**Tujuan**:untuk mendekati nilai konstanta $e$ melalui deret Mac Laurin.
$$
e^x= \sum_{n=0}^{\infty} \frac{x^n}{n!}
$$

Jadi Kita akan menghitung:

$$
e^x=\frac{x^0}{0!} +\frac{x^1}{1!} + \ldots + \frac{x^n}{n!}
$$


```python
from __future__ import division
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
def hitung_error(t1,t0):
    error =abs(t1 - t0) / t1 * 100
    return error
```

```python
def hitung_deret_v1(max_iter ,x ):
    faktorial = 1
    hasil = 0
    
    
    for n in range(max_iter + 1):
        if(n !=0 ):
            faktorial =faktorial * n
        hasil =hasil + (x ** n) / faktorial
    return hasil
```

```python
#hasil sebenarnya
print (np.exp(1))

#menghitung e^1
print (hitung_deret_v1( x=1, max_iter=5))

print (hitung_deret_v1( x=1, max_iter=13))
```

```python
def hitung_deret_v2(max_iter ,x ):
    faktorial = 1
    hasil = 0
    
    arr_hasil=[]
    
    for n in range(max_iter + 1):
        if(n !=0 ):
            faktorial =faktorial * n
        hasil =hasil + (x ** n) / faktorial
        arr_hasil.append(hasil)
    return (hasil,arr_hasil)
```

```python
(hasil,arr_hasil)=hitung_deret_v2(x=1,max_iter=10)
```

```python
def hitung_deret_v4(x ,threshold):
    
    faktorial = 1
    hasil = 0
    n=0
    arr_hasil=[]
    arr_error=[]
    error=100
    while (error > threshold):
        if(n !=0 ):
            faktorial =faktorial * n
            hasil =hasil + (x ** n) / faktorial
            arr_hasil.append(hasil)
            error=hitung_error(arr_hasil[n],arr_hasil[n-1])
            print(error)
            
            arr_error.append(error)
            
        else :
            error=hitung_error(arr_hasil[n],arr_hasil[n-1])
            arr_hasil.append(hasil)
            n+1
    
    return (hasil,arr_hasil,n)
```

```python
(hasil,arr_hasil,N)=hitung_deret_v4( x=1 , threshold= 0.05)
print ("Iterasi mencapai konvergensi pada suku ke-%d" % (N-1))
print "Error = %.4f%%" % (arr_error[len(arr_error)-1])

plt.subplot(211)
plt.plot(range(N),arr_hasil,'b,-')
plt.ylabel('Nilai e')

# plt.subplot(212)
# plt.plot(range(1,N),arr_error,'r--')
# plt.xlabel('Iterasi')
# plt.ylabel('Error')
# plt.show()
```
